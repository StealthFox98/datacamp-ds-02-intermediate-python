# 2.2 Dictionaries and Pandas

## Dictionaries
Dictionaries are used to store data values in key:value pairs.

### Dictionaries vs. lists
Aspects of a list
- A sequence of values indexed by a sequence of numbers.
- Used for a collection of values where the order matters, or for selecting entire subsets.

Aspects of a dictionary
- A sequence of values indexed by unique keys, that can be immutable type.
- Chosen if you want a lookup table with unique keys.

```python
pop = [30.55, 2.77, 39.21]
countries = ['afghanistan', 'albania', 'algeria']
```

```python
# Example of a list approach
ind_alb = countries.index('albania')
print(ind_alb)
print(pop[ind_alb])
```

```python
# Example of a list approach
world = dict(zip(countries,pop))
world
```

```python
world['albania']
```

```python
print(world.keys()) # keys of a dict
print(world.values()) # values of a dict
```

### More about dictionaries
- Keys have to be "immutable" objects (cannot be changed once they are created, like: int, float, str, bool)
- To add a new set of information, use:

        dict['key'] = value

- You can change values by using:

        dict['key'] = value_new

- To delete a key:value pair, write:

        del(dict['key'])

- To find out if a key is in the dict

        'key' in dict

```python
world['sealand'] = 0.00027
world
```

```python
'sealand' in world # If it exists in dict, value will be True
```

```python
world['sealand'] = 0.000028
world
```

```python
del(world['sealand'])
world
```

```python
print('sealand' in world)
```

### for loop
Used for iterating over a sequence (list, tuple, dictionary, set, string).

            expression

```python
fam = [1.73, 1.68, 1.71, 1.89]
fam
```

# 2.5 Random numbers

```python
np.random.rand() # Pseudo-random numbers
```

```python
np.random.seed(123) # Same seed, same random numbers.
np.random.rand() # Same random numbers ensures reproductibility
```

```python
# Simulating a coin toss
np.random.seed(123)
coin = np.random.randint(0,2) # Randomly generate 0 or 1
print(coin)
if coin == 0:
    print("heads")
else:
    print("tails")
```

```python
# Heads or tails
# range(start, stop, step) returns a sequence of numbers
np.random.seed(123)
outcomes = []
for x in range(10):
    coin = np.random.randint(0,2)
    if coin == 0:
        outcomes.append("heads")
    else:
        outcomes.append("tails")

print(outcomes)
```

```python
# Heads or tails: Random walk
np.random.seed(123)
tails = [0]
for x in range(10):
    coin = np.random.randint(0,2)
    tails.append(tails[x] + coin)
print(tails)
```

## Distribution

```python
# Heads or tails: Random walk, n runs
np.random.seed(123)
final_tails = []
for x in range(10000):
    tails = [0]
    for x in range(10):
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])
print(final_tails)
```

```python
import matplotlib.pyplot as plt
plt.hist(final_tails, bins=10)
plt.show()
```
